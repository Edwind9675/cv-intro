```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
I = cv2.imread('test_image3.jpg')
cap = cv2.VideoCapture('AUV_Vid.mkv')
from dt_apriltags import Detector
import lane_detection

```

```python
def image_line_detect(I):
    img0 = I.copy()
    gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY) # convert to grayscale
    edges = cv2.Canny(gray, 470, 301, apertureSize=5) # detect edges
    lines = cv2.HoughLinesP(
                edges,
                1,
                np.pi/180,
                100,
                minLineLength=300,
                maxLineGap=10,
        ) # detect lines

    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(img0, (x1, y1), (x2, y2), (0, 255, 0), 2)
        print ((y2-y1)/(x2-x1))

    #image_line_detect(img)
    plt.imshow(img0)

```

```python

image_line_detect(cv2.imread('test_image3.jpg'))
```
